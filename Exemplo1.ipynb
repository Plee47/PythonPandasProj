```python
revendas = comb_df["Revenda"]
display(revendas)
```

```python
produtos_df = comb_df[['Revenda', 'Municipio', 'Produto', 'Valor de Venda']]
display(produtos_df)
```

```python
display(comb_df.loc[1:5])
```

```python
gaso_df = comb_df.loc[comb_df['Produto'] =='GASOLINA']
display(gaso_df)

display(gaso_df['Valor de Venda'].max())
```

```python
display(comb_df[['Revenda', 'Municipio', 'Valor de Venda']].max())
```

```python
gaso_df = comb_df.loc[(comb_df['Produto'] == 'GASOLINA') & (comb_df['Municipio'] == 'INDAIATUBA')]
display(gaso_df)
```

```python
gaso_df = comb_df.loc[(comb_df['Produto'] == 'GASOLINA') & (comb_df['Municipio'] == 'INDAIATUBA'), 
                                         ['Municipio', 'Revenda', 'Data da Coleta', 'Produto', 'Valor de Venda']]
display(gaso_df)
```

```python
gaso_df['Imposto'] = gaso_df['Valor de Venda'] * 0.10
display(gaso_df)
```

```python
gaso_df = gaso_df.drop("Imposto", axis=1)
display(gaso_df)
```

```python
gaso_df = gaso_df.dropna(how='all')
display(gaso_df)
```

```python
gaso_df = gaso_df.dropna(how='all', axis=1)
display(gaso_df)
```

```python
gaso_df['Imposto'] = gaso_df['Valor de Venda'] * 0.10
gaso_df['Imposto'] = gaso_df['Imposto'].fillna(gaso_df['Imposto'].mean())
display(gaso_df)
```

```python
media_por_prod = comb_df[['Produto', 'Valor de Venda']].groupby('Produto').mean().round(2)
display(media_por_prod)
```

```python
comb_df['Ativo'] = True
print(comb_df.info())
display(comb_df.head)
```

```python
media_por_prod.to_excel("media_por_prod.xlsx", sheet_name = 'Media de valor dos produtos', index = False)
```
